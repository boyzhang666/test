```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import seaborn as sns  # Using seaborn just to load the iris dataset

# Load the iris dataset
iris = sns.load_dataset('iris')
iris
```

```python
def tu(iris):
    length = iris.shape[1]
    # Create a 4x4 subplot grid
    fig = make_subplots(rows=length, cols=length,
                        # subplot_titles=('sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)'),
                        shared_xaxes=False, shared_yaxes=False,
                        horizontal_spacing=0.03,  
                        vertical_spacing=0.03)

    # Scatter plots and histograms
    features = iris.columns.tolist()

    # 获取每一个特征的最大值和最小值
    ranges = {feat: [iris[feat].min()-1, iris[feat].max()+1] for feat in features}

    for i, feat_x in enumerate(features):
        for j, feat_y in enumerate(features):
            if i == j:  # Diagonal - Histogram
                fig.add_trace(go.Histogram(x=iris[feat_x], name=feat_x, nbinsx=20,), row=i+1, col=j+1)
                if i == 0:
                    fig.update_yaxes(range=ranges[feat_x], row=i+1, col=j+1, 
                                    title=feat_x 
                                    )  # 仅为散点图更新轴范围
                elif i == 3:
                    fig.update_xaxes(range=ranges[feat_x], row=i+1, col=j+1, 
                                    title=feat_x 
                                    )  # 仅为散点图更新轴范围
            else:  # Scatter plot
                fig.add_trace(go.Scatter(x=iris[feat_x], y=iris[feat_y], mode='markers', name=feat_y), row=i+1, col=j+1)
                fig.update_yaxes(range=ranges[feat_y],
                                title=feat_y if j == 0 else None, 
                                title_standoff=5,  # adjust title distance
                                row=i+1, col=j+1,)  # 仅为散点图更新轴范围
                fig.update_xaxes(range=ranges[feat_x], 
                                title=feat_x if i == 3 else None,
                                title_standoff=15,  # adjust title distance
                                row=i+1, col=j+1, )  # 仅为散点图更新轴范围

    # Update layout
    fig.update_layout(title_text="Pair plot of the Iris dataset", showlegend=False, width=1200, height=1200)
    fig.update_traces(marker=dict(size=5), selector=dict(mode='markers'))

    fig.show()
```
